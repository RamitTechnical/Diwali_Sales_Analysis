# EDA(Explanatory Data Analysis)

### Using the number of males and females using countplot

```python
plt.figure(figsize=(4,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
```

```python
df.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
ax=sns.barplot(x="Gender",y="Amount",data=sales_gen)
ax.bar_label(ax.containers[0])
#or
#ax=sns.barplot(data=df,x="Gender",y="Amount",estimator=np.sum)
#ax.bar_label(ax.containers[0])
```

### Finding: Most of the buyers are females and even the purchasing power of females are greater than men

# Age

```python
#Finding the number of males and females in different Age Group
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
age_amount=df.groupby("Age Group",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
#or
#sns.barplot(data=df,x="Age Group",y="Amount",estimator=np.sum)
```

### Finding the total amount of sales done by Age Group

```python
sns.barplot(data=age_amount,x="Age Group",y="Amount")
```

### Finding:It is concluded that most of the buyers are from Age Group:25-35 years 

# State

### Total number of orders from TOP 10 states

```python
plt.figure(figsize=(15,5))
orders_state=df.groupby("State",as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.barplot(data=orders_state,x="State",y="Orders")
```

### Total amount/Sales from Top 10 Sates

```python
plt.figure(figsize=(15,5))
orders_state=df.groupby("State",as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sns.barplot(data=orders_state,x="State",y="Amount")
```

### Finding: We can see that most of the orders & total sales/amount are from Uttar Pradesh,Maharashtra and Karnataka respectively.

# Martial Status

### Checking the marital_status

```python
ax=sns.countplot(data=df,x="Marital_Status")
ax.bar_label(ax.containers[0])
```

```python
#Want to know the amount spend by married and unmarried people on Gender basis
marital_sales=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
ax=sns.barplot(data=marital_sales,x="Marital_Status",y="Amount",hue="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

### Finding: We can see that most of the buyers are married(women) 

# Occupation

### Count in different Occupation

```python
plt.figure(figsize=(20,5))
sns.countplot(data=df,x="Occupation")
```

### Purchasing Power of different sector

```python
plt.figure(figsize=(20,5))
Occupation_amount=df.groupby("Occupation",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
ax=sns.barplot(data=Occupation_amount,x="Occupation",y="Amount")
ax.bar_label(ax.containers[0])
plt.show()
```

### Finding: We can see that most of the buyers are from IT,Heathcare,Aviation,Banking

# Product Category

### Count of the different Product Category

```python
plt.figure(figsize=(20,5))
sns.countplot(data=df,x="Product_Category")
```

```python
plt.figure(figsize=(20,5))
Product_Category_amount=df.groupby("Product_Category",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
ax=sns.barplot(data=Product_Category_amount,x="Product_Category",y="Amount")
```

### Finding: We can see that most of the sold products are from Food,Clothing and Electronics category
